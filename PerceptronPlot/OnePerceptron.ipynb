```python
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib inline
%matplotlib widget
```

```python
def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1/(1 + np.exp(-x))

def hard_limit(x: np.ndarray) -> np.ndarray:
    return np.where(x >= 0, 1, -1)

def radial_basis(x: np.ndarray) -> np.ndarray:
    """
    Source: https://www.simplilearn.com/tutorials/machine-learning-tutorial/what-are-radial-basis-functions-neural-networks

    - A Radial Basis Function (RBF) is a real-valued function whose value depends only on the distance from the origin or some other center point.
    - RBF are used as an activation functions, particularly in RBF networks.
    - A common form of RBF is the Gaussian function: φ(x) = exp(-β||x - c||²)


    """
    
    return np.exp(-x**2)
```

```python
# number of points
n = 100

# generate some random points
x1 = np.linspace(-2, 2, n)
x2 = np.linspace(-2, 2, n)

# calculate the function
# y = sigma(-4.79 * x1 + 5.9 * x2 - 0.93)

def y (x1: np.ndarray, x2:np.ndarray, sigma:object) -> np.ndarray:
    return sigma(-4.79 * x1 + 5.9 * x2 - 0.93)
```

```python
##### creating 3d plot
# https://www.educba.com/seaborn-3d-plot/

sigma = sigmoid  # sigmoid | hard_limit | radial_basis

# creating all possible combinations of x1 and x2 using meshgrid
plot1, plot2 = np.meshgrid(x1, x2)
plot3 = y(plot1, plot2, sigma=sigma)

# creating the plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surface = ax.plot_surface(plot1, plot2, plot3, cmap="twilight_shifted")

# setting labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title(f'Sigmoid function - {n=}')

# saving the plot
plt.savefig(f'{sigma.__name__}-{n}.svg', dpi=300)
plt.savefig(f'{sigma.__name__}-{n}.png', dpi=300)

plt.show()
```
